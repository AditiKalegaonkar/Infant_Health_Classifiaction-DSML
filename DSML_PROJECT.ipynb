## Encoding the Data

```python
# Converting the string values to comparable float and int values to train the model and do the predictions

df["BirthAsphyxia"].replace('yes',1, inplace = True)
df["BirthAsphyxia"].replace('no',0,inplace = True)

df["HypDistrib"].replace('Equal',1,inplace = True)
df["HypDistrib"].replace('Unequal',0,inplace = True)

df["HypoxiaInO2"].replace('Severe',3, inplace = True)
df["HypoxiaInO2"].replace('Moderate',2,inplace = True)
df["HypoxiaInO2"].replace('Mild',1,inplace = True)

df["CO2"].replace('High',3, inplace = True)
df["CO2"].replace('Normal',2,inplace = True)
df["CO2"].replace('Low',1,inplace = True)


df["Grunting"].replace('yes',1, inplace = True)
df["Grunting"].replace('no',0,inplace = True)


df["LVHreport"].replace('yes',1, inplace = True)
df["LVHreport"].replace('no',0,inplace = True)

df["LowerBodyO2"].replace('5-12',3, inplace = True)
df["LowerBodyO2"].replace('<5',2,inplace = True)
df["LowerBodyO2"].replace('12+',1,inplace = True)

df["RUQO2"].replace('5-12',3, inplace = True)
df["RUQO2"].replace('<5',2,inplace = True)
df["RUQO2"].replace('12+',1,inplace = True)

df["CO2Report"].replace('<7.5',1, inplace = True)
df["CO2Report"].replace('>=7.5',0,inplace = True)


df["ChestXray"].replace('Oligaemic',1, inplace = True)
df["ChestXray"].replace('Normal',2,inplace = True)
df["ChestXray"].replace('Plethoric',3,inplace = True)
df["ChestXray"].replace('Asy/Patch',4,inplace = True)
df["ChestXray"].replace('Grd_Glass',5,inplace = True)

df["LungParench"].replace('Abnormal',10, inplace = True)
df["LungParench"].replace('Normal',0,inplace = True)
df["LungParench"].replace('Congested',5,inplace = True)

df["Age"].replace('0-3_days',0, inplace = True)
df["Age"].replace('4-10_days',5,inplace = True)
df["Age"].replace('11-30_days',10,inplace = True)


df["LungFlow"].replace('High',10, inplace = True)
df["LungFlow"].replace('Normal',5,inplace = True)
df["LungFlow"].replace('Low',0,inplace = True)

df["Sick"].replace('yes',1, inplace = True)
df["Sick"].replace('no',0,inplace = True)

df["XrayReport"].replace('Oligaemic',1, inplace = True)
df["XrayReport"].replace('Normal',2,inplace = True)
df["XrayReport"].replace('Plethoric',3,inplace = True)
df["XrayReport"].replace('Asy/Patchy',4,inplace = True)
df["XrayReport"].replace('Grd_Glass',5,inplace = True)

df["LVH"].replace('yes',1, inplace = True)
df["LVH"].replace('no',0,inplace = True)

df["GruntingReport"].replace('yes',1, inplace = True)
df["GruntingReport"].replace('no',0,inplace = True)

df["Disease"].replace("TGA",1,inplace = True)
df["Disease"].replace("Fallot",2,inplace = True)
df["Disease"].replace("PFC",3,inplace = True)
df["Disease"].replace("PAIVS",4,inplace = True)
df["Disease"].replace("TAPVD",5,inplace = True)
df["Disease"].replace("Lung",5,inplace = True) 

df
```

## Cross Validation

```python
x = df.drop(["Sick"], axis = 1)
y = df["Sick"]                 
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20, random_state = 0)
```

## Building the Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators = 10)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test) 
```

```python
df1 = pd.DataFrame({"Actual": y_test,"Predicted": y_pred})
df1
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
confusion_matrix(y_test,y_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
print(classification_report(y_test,y_pred))
```

```python
RFC = accuracy_score(y_test,y_pred)
```

## Building the model for KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 60)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
KNN = accuracy_score(y_test,y_pred)
```

## Feature Selection

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
sfs = SequentialFeatureSelector(knn,direction = "backward") 
```

```python
sfs.fit(x_train,y_train)
```

```python
sfs.get_support()
```

```python
x = x.iloc[:,sfs.get_support()]
```

```python
knn.fit(x,y)
```

```python
y_pred = knn.predict(x)
accuracy_score(y,y_pred)
```

```python
KNN1 = knn.score(x,y)
```

## Using hyperparameter tuning for increasing accuracy

```python
params = {
    "n_neighbors": [100,200,300,400],
    "weights" : ["uniform","distance"],
    "algorithm":["auto","ball_tree","kd_tree","brute"]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid =  GridSearchCV(estimator =KNeighborsClassifier(),
                    param_grid = params,
                    scoring = "accuracy",
                    cv = 5)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_estimator_
```

```python
y_pred = grid.predict(x_test)
```

```python
KNN2 = accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred));
```
